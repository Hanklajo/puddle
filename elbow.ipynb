```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# สร้างข้อมูล x และ y โดยใช้ NumPy
x = np.arange(1, 501)  # สร้าง array ของ x ที่มีค่าตั้งแต่ 1 ถึง 100
y = np.random.randint(100, 201, size=500)  # สร้าง array ของ y ที่มีค่าสุ่มในช่วง 100-200 โดยมีขนาดเท่ากับ 100

# นำข้อมูล x และ y มาต่อกันเป็นจุดในระบบพิกัด 2 มิติ
X = np.column_stack((x, y))

# ทดสอบจำนวน cluster ตั้งแต่ 1 ถึง 10
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# พล็อตกราฟ Elbow Method
plt.plot(range(1, 10), wcss)
plt.title('Elbow Method')
plt.xlabel('จำนวนของ Cluster')
plt.ylabel('WCSS')
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x = np.arange(1,501)
y = np.random.randint(100,601,size=500)
Data = np.column_stack((x,y))
```

```python

wcss = []
for idx in range(1,10):
    kmeans = KMeans(n_clusters=idx , init="k-means++", max_iter = 300 , n_init=10 )
    kmeans.fit(Data)
    wcss.append(kmeans.inertia_)

```

```python
for idx in range (len(wcss)):
    print(wcss[idx],'------K=', idx+1)
```
