```python
import numpy as np
import matplotlib.pyplot as plt


```

```python

X = np.random.rand(100)
Y = np.random.rand(100)

K = 3 
centroid_x = np.random.rand(K)
centroid_y = np.random.rand(K)
plt.plot(X,Y,'o')
plt.plot(centroid_x,centroid_y,'o')
distance = np.sqrt((X[:, np.newaxis]-centroid_x)**2 + (Y[:, np.newaxis]-centroid_y)**2)
cluster = np.argmin(distance , axis=1)


for idx in range (len(X)):
    print('ค่าที่',idx,distance[idx],' กลุ่ม  ',cluster[idx],'---' ,'X' ,X[idx] ,'Y',Y[idx])

iteration = 500

for idx in range(iteration):
    distances = np.sqrt((X[:, np.newaxis] - centroid_x)**2 + (Y[:, np.newaxis] - centroid_y)**2)
    clusters = np.argmin(distances, axis=1)
    new_centroid_x = np.array([X[clusters == i].mean() for i in range(K)])
    new_centroid_y = np.array([Y[clusters == i].mean() for i in range(K)])


    if np.allclose(centroid_x, new_centroid_x) and np.allclose(centroid_y, new_centroid_y):
        print('รอบ',idx)
        break
    centroid_x, centroid_y = new_centroid_x, new_centroid_y
    
```
