```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = 20
noise = 2*np.random.rand(num)+1
x = np.arange(num)
x1 = np.arange(num)+0.5
y = np.arange(num) -noise
np.random.rand(num)
```

```python
plt.plot(x,y, 'o')
```

```python
def h(theta0, theta1, x):
    return theta0+theta1*x
```

```python
theta0 = np.random.rand(1)
theta1 = np.random.rand(1)
theta2 = np.random.rand(1)
```

```python
lr = 0.00001
epoch = 1000
```

```python
for i in range(epoch):
    error = h(theta0,theta1,x)-y
    theta0 = theta0 - lr*(1/num) * np.sum(error)
    theta1 = theta1 - lr*(1/num) * np.sum(error*x)
```

```python
y_pred = h(theta0,theta1,x)
print(y_pred)
```

```python
plt.plot(x,y,'o')
plt.plot(x,y_pred)
```
